```
import importlib
importlib.import_module('mpl_toolkits').__path__
```

```

from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
fig = plt.figure()
ax = plt.axes(projection='3d')
```

```
ax = plt.axes(projection='3d')
zline = np.linspace(0, 20, 1500)
xline = np.sin(zline)
yline = np.cos(zline)
ax.plot3D(xline, yline, zline, 'gray')
zdata = 20 * np.random.random(100)
xdata = np.sin(zdata) + 0.2 * np.random.randn(100)
ydata = np.cos(zdata) + 0.2 * np.random.randn(100)
ax.scatter3D(xdata, ydata, zdata, c=zdata, cmap='Greens')
```

```
def f(x, y):
    return np.sin(np.sqrt(x ** 2 + y ** 2))

X = np.linspace(-1, 5, 60)
Y = np.linspace(-2, 8, 100)

x, y = np.meshgrid(X, Y)
z = f(x, y)
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(x,y,z, 200, cmap='binary')
ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')
```

```
theta = 2 * np.pi * np.random.random(10000)
r = 6 * np.random.random(10000)

x = np.ravel(r * np.sin(theta))
y = np.ravel(r * np.cos(theta))
z = f(x, y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, cmap='viridis') 
```
